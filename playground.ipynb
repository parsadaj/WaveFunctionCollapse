```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from WFC import WaveFunctionCollapse, WaveFunctionCollapseVisualizer
from matplotlib.colors import ListedColormap
```

```python
# Example Usage
input_image = np.random.random_integers(0,2, (4,4))
input_image = [[0,0,0,0],
               [0,1,1,1],
               [0,1,2,1],
               [0,1,1,1]]

cmap = ListedColormap(['grey', 'black', 'red'])

sns.heatmap(input_image, annot=True,cbar=False, cmap=cmap, vmin=0, vmax=2)
```

```python


wfc = WaveFunctionCollapse(input_image, pattern_size=2, grid_size=(12,12))
output_image = wfc.run()

# Example usage:
# Assuming you have a grid size, a list of observations, and a mapping from patterns to numbers
observations = wfc.observations[:10]

pattern_to_number = wfc.pattern_to_number
visualizer = WaveFunctionCollapseVisualizer(grid_size=(12,12), observations=observations, pattern_to_number=pattern_to_number, pattern_size=2, cmap=cmap, adjacency_rules=wfc.adjacency_rules)
visualizer.visualize_patterns()
```

```python
for pat in wfc.patterns:
    visualizer.visualize_adjacency(pat)
```
